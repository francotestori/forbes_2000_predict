```python
# We filter all companies with null numerical values and we re calculate % columns
df = df[(df.market_value.notnull()) & (df.profits.notnull()) & (df.assets.notnull()) & (df.sales.notnull())]

df['profits_%_assets'] = df.profits / df.assets
df['profits_%_sales'] = df.profits / df.sales
df = df.fillna({'profits_%_sales': 0.0})
```
